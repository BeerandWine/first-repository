```python
# load_digits
import sklearn
print(sklearn.__version__)
import pandas as pd

from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

digits = load_digits()
digits.keys()
digits_data = digits.data # Feature Data 지정하기
digits_label = digits.target # Label Data 지정하기
digits.target_names  # Target Names 출력해보기
digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names)
digits_df['label'] = digits.target
digits_df # 데이터 Describe 해 보기
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data,
                                                    digits_label,
                                                    test_size=0.2,
                                                    random_state=7)

# load_digits > decision_tree 모델 학습 & 예측
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# load_digits > RandomForest 모델 학습 & 예측
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=32)
# 데이터를 분리할 때 random_state와, 모델을 학습시킬 때 random_state는 어떻게 다를까?
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# load_digits > Support Vector Machine 모델 학습 & 예측
from sklearn import svm
svm_model = svm.SVC()
#과정중에는 "이 모델이 어떤 타입인지 확인하는 과정"이 왜 추가되어 있을까?
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# load_digits > Stochastic Gradient Descent 모델 학습 & 예측
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# load_digits > Rogistic Regression 모델 학습 & 예측
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))
```
